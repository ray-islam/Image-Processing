```python
from ctgan import CTGAN
from ctgan.synthesizers.ctgan import Generator
```

```python
data = pd.read_csv('./GenAI.csv')
```

```python
data
```

```python
categorical_features = ['species']
```

```python
continuous_cols = ['sepel_length', 'sepal_width', 'petal_length','petal_width']
```

```python
from ctgan import CTGAN

ctgan = CTGAN(verbose=True)
ctgan.fit(data, categorical_features, epochs = 50000)
```

```python
samples = ctgan.sample(50000)
```

```python
samples
```

```python
# round continuous data to 1 decimal place
# assuming generated_data is a pandas DataFrame containing the generated data
continuous_cols = ['sepel_length', 'sepal_width','petal_length','petal_width' ]

for col in continuous_cols:
    samples[col] = samples[col].apply(lambda x: round(x, 1))
```

```python
samples
```

```python
# Export DataFrame to CSV file
samples.to_csv('Synthetic_IRIS_Data.csv', index=False)
```

```python
# Convert DataFrame to a string
data_str = samples.to_string(index=False)

# Export data to text file
with open('Synthetic_IRIS_Data.txt', 'w') as f:
    f.write(data_str)
```
